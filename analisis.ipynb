```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.clf()
```

```python
## mengambil data contoh
raw_data = pd.read_csv("omzet_tanaman.csv")
print(raw_data)
```

```python
print(raw_data["omzet"].mean())
print(raw_data["omzet"].median())
print(raw_data["omzet"].value_counts())
print(raw_data["omzet"].quantile([0.25,0.5,0.75]))
```

```python
plt.figure()
raw_data['omzet'].hist()
plt.title('omzet', size=20)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20,10))
raw_data['lokasi_penjualan'].hist()
plt.title('Histogram pendapatan', size=14)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.scatter(x='lokasi_penjualan', y='omzet', data=raw_data)
plt.title('Diagram Scatter', size=14)
plt.xlabel('Lokasi Penjualan')
plt.ylabel('Omzet')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.bar(x='lokasi_penjualan', height='omzet',data=raw_data)
plt.title('Histogram Hubungan Lokasi Penjualan dengan Omzet', size=14)
plt.xlabel('Lokasi Penjualan')
plt.ylabel('Omzet')
plt.tight_layout()
plt.show()
```
